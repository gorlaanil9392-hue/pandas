```
# LINE GRAPH ?

import matplotlib.pyplot as plt
x=[2,3,4]
y=[3,7,5]
plt.plot(x,y)
plt.show()
```

```
# LINE GRAPH ?

import matplotlib.pyplot as plt
x=[12,34,65]
y=[10,30,43]
plt.plot(x,y,marker='*')
plt.show()
```

```
# using LINE graph ?

import matplotlib.pyplot as plt
x=[10,20,30]
y=[23,59,37]
plt.plot(x,y,marker='*')
plt.title('line Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```
# using BAR graph ?

import matplotlib.pyplot as plt
days=['sun','mon','tue','wed','thur','fri','sat']
data=[10,20,30,40,50,60,70]
plt.bar(days,data)
plt.title('Bar graph')
plt.xlabel('days')
plt.ylabel('data')
plt.show()

```

```
# using horizental bar graph ?

import matplotlib.pyplot as plt
data=[10,20,30,40,50,60,70]
plt.barh(days,data)
plt.title('Bar graph')
plt.xlabel('days')
plt.ylabel('data')
plt.show()
```

```
# using width bar graph

import matplotlib.pyplot as plt
days=['sun','mon','tue','wed','thur','fri','sat']
data=[10,20,30,40,50,60,70]
plt.bar(days,data,width=0.1)
plt.title('Bar graph')
plt.xlabel('x-days')
plt.ylabel('y-data')
plt.show()
```

```
# using 2 lines graph ?

import matplotlib.pyplot as plt
x=[3,5,7]
y=[2,6,8]
p=[10,20,30]
plt.plot(x,y,marker='*')
plt.plot(x,p,marker='*')
plt.title('2 line graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```
# using bar graph ?

import matplotlib.pyplot as plt
data=[10,12,56,78,34,17,97,56,20,30,40,50,60,70,80,90]
plt.hist(data,bins=20)
plt.title('histogram plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```
# using pie chart?

import matplotlib.pyplot as plt
frinds=['anil','suresh','vamsi']
marks=[90,30,50]
plt.pie(marks,labels=frinds)
plt.show()
```

```
# using pie chart?

import matplotlib.pyplot as plt
friends=['anil','suresh','karthik']
marks=[70,60,30]
c=['g','m','b']
plt.pie(marks,labels=friends,colors=c,shadow="true",explode=(0.1,0.1,0.1),autopct='%1.1f%%')
plt.title('Buddies')
plt.axis('equal')
plt.legend()
plt.show()
```

```
# using scatter plot?

import matplotlib.pyplot as plt
data=[10,23,54,65,87]
places=[3,5,2,6,7]
plt.scatter(data,places)
plt.show()
```

```
# using scattor plot ?

import matplotlib.pyplot as plt
area=[2,4,3,5,7]
flats=[12,14,13,18,19]
plt.scatter(area,flats)
plt.title('Township')
plt.xlabel('area')
plt.ylabel('flats')
plt.show()
```

```
# using BOX PLOT chart ?

import matplotlib.pyplot as plt
a=[3,4,7,8,5]
plt.boxplot(a)
plt.show()

```

```
# BOX PLOT

import matplotlib.pyplot as plt
a=[10,20,30,40,50,60,200]
plt.boxplot(a)
plt.show()
```
